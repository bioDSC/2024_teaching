# Lesson 13, Conditionals

````python
# if statement
    # properly called "conditional statement"    
    
    # controls whether code gets executed
    
mass = 3.54
if mass > 3.0:
    print(mass, 'is large')

mass = 2.07
if mass > 3.0:
    print (mass, 'is large')
    
# structure of if statement:
    # first line opens with "if" statement
    # body is indented
    
# e.g. useful when looking at multiple cases

masses = [3.54, 2.07, 9.22, 1.86, 1.71]
for m in masses:
    if m > 3.0:
        print(m, 'is large')
          
        
    
````

````python
# Use "else" 
    # what to do when statement not true

masses = [3.54, 2.07, 9.22, 1.86, 1.71]
for m in masses:
    if m > 3.0:
        print(m, 'is large')
    else:
        print(m, 'is small')
````

````python
# elif 
    # to specify another test
    # for multiple options
    # short for "else if"    
    # else comes at the end
        # "catch all"

masses = [3.54, 2.07, 9.22, 1.86, 1.71]
for m in masses:
    if m > 9.0:
        print(m, 'is HUGE')
    elif m > 3.0:
        print(m, 'is large')
    else:
        print(m, 'is small')
        
        
# note that order is very important

# Adressing potential misconception

velocity = 10.0
if velocity > 20.0:
    print('moving too fast')
else:
    print('adjusting velocity')
    velocity = 50.0 
        # now velocity is 50
        # but we don't go "back" to first statement
        # "moving too fast" won't be executed
        
````

````python
# Example with slightly more complicated loop

# Adjustment in steps
velocity = 10.0
for i in range(5): # execute the loop 5 times
    print(i, ':', velocity)
    
    if velocity > 20.0:
        print('moving too fast')
        velocity = velocity - 5.0
    else:
        print('moving too slow')
        velocity = velocity + 10.0
        
print('final velocity:', velocity)


# Go through each loop iteration, explaining what value it will have
````
