```python
import numpy as np

import matplotlib.pyplot as plt

# Generate data for Gaussian curve
x = np.linspace(-5, 5, 100)
y = (1/np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Now draw some numbers from the same gaussian distribution
n = 12
y_samples = np.random.normal(0, 1, n)-.5
# calculate histogram
hist, bins = np.histogram(y_samples, bins=30, density=True)

# Plot Gaussian curve
plt.plot(x, y, linewidth=5, color='darkred')
plt.hist(y_samples, bins=30, color='grey', alpha=0.5, density=True)
plt.ylabel('Amount of people')
plt.xlabel('Appropriate learning rate')
# turn off x axis
plt.xticks([])
plt.yticks([])
plt.box(False)
# set font to 15 pts
plt.rc('font', size=25)
plt.ylim([-.2, np.max(hist)*1.2])
plt.show()
```

```python
# create a dataframe with columns cond1 and cond2 and add some random values in 10 rows, where the average of cond1 is 1.5x higher than cond2

import pandas as pd
from scipy.stats import ttest_ind

# Create a dataframe with columns cond1 and cond2
data = {
    'cond1': np.random.normal(3, 1, 10),
    'cond2': np.random.normal(2, 1, 10)
}
df = pd.DataFrame(data)
print(df)

# compare the columns cond1 and cond2 with a simple t-test
# Perform t-test
t_stat, p_value = ttest_ind(df['cond1'], df['cond2'])
print(f"T-statistic: {t_stat}, P-value: {p_value}")

```
