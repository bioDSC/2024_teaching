````python
# matplotlib most widely used
    # or sub library "pyplot"
    # will render inline by default

import matplotlib.pyplot as plt

# simple plot

time = [0, 1, 2, 3]
position = [0, 100, 200, 300]

plt.plot(time, position)
plt.xlabel('Time (hr)')
plt.ylabel('Position (km)')
````

````python
# Can also plot directly from pandas

import pandas as pd

data = pd.read_csv('/Users/m.wehrens/Data_UVA/2024_teaching/2025-03-gapminder/data/gapminder_gdp_oceania.csv', \
    index_col='country')

# search and replace
    # keep year
    # remove rest

years = data.columns.str.replace('gdpPercap_', '')
    # MW:
    # "str" is a method which holds more methods 
        # related to string operations
        # str is also part of python standard library
        # stand-alone example: str.replace('hoi', 'oi', '') 

# convert to int
# put back in dataframe
data.columns = years.astype(int)

data.loc['Australia'].plot()
````

````python
# Select and transform data, then plot it

# Default is plotting rows as x-axis
    # note to self: consistent with tidyverse

# So, to plot versus year, we need transpose:
data.T.plot()
plt.ylabel('GDP per capita')
    # columns are automatically one line
    
# MW:
print(data.T.head())
````

````python
# Many styles are available

# Example
plt.style.use('ggplot')
data.T.plot(kind='bar')
plt.ylabel('GDP per capita')

# use google to tweak
# https://matplotlib.org/stable/gallery/ for nice examples
# (also seaborn)
````
